```python
# dataframe for plotting companies per country
df_graph_country = df_graph[['company_name','country']]
```

```python
df_graph_country.drop_duplicates(inplace=True)
```

```python
# make a count of companies in each country with axes label, title and value label
df_graph_country['country'].value_counts().plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Companies')
plt.title('Number of Companies in Each Country')
for i in range(len(df_graph_country['country'].value_counts())):
    plt.text(i, df_graph_country['country'].value_counts()[i], df_graph_country['country'].value_counts()[i], ha = 'center')
plt.show()
```

```python
# dataframe for plotting companies per top 10 city and state
df_graph_city = df_graph[['company_name','city']]
df_graph_city['city'] = df_graph_city['city'].str.upper()
df_graph_city.drop_duplicates(inplace=True)
df_graph_state = df_graph[['company_name','state']]
df_graph_state['state'] = df_graph_state['state'].str.upper()
df_graph_city.drop_duplicates(inplace=True)
df_graph_state.drop_duplicates(inplace=True)

df_modified = df_graph_city.copy()

# Convert city names to uppercase for consistency
df_modified['city'] = df_modified['city'].str.upper()

# Define a mapping of patterns to standard city names
pattern_mapping = {
    '.*MUM.*': 'Mumbai',
}

# Apply the pattern matching and replacement
for pattern, replacement in pattern_mapping.items():
    df_modified.loc[df_modified['city'].str.contains(pattern, regex=True), 'city'] = replacement

fig, ax = plt.subplots(1, 2, figsize=(20, 10))

df_modified['city'].value_counts().head(10).plot(kind='bar', ax=ax[0])
ax[0].set_xlabel('City')
ax[0].set_ylabel('Number of Companies')
ax[0].set_title('Number of Companies in Each City')
for i, value in enumerate(df_modified['city'].value_counts().head(10)):
    ax[0].text(i, value, value, ha='center', va='bottom')

df_graph_state['state'].value_counts().head(10).plot(kind='bar', ax=ax[1])
ax[1].set_xlabel('State')
ax[1].set_ylabel('Number of Companies')
ax[1].set_title('Number of Companies in Each State')
for i, value in enumerate(df_graph_state['state'].value_counts().head(10)):
    ax[1].text(i, value, value, ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
data = {'city': ['Mumbai', 'Chennai', 'Bangalore', 'Pune', 'Hyderabad', 'Mumbai', 'Delhi', 'Chennai', 'Kolkata', 'Mumbai', 'Mumbai', 'Mumbai', 'Chennai', 'Pune'],
        'state': ['Maharashtra', 'Tamil Nadu', 'Karnataka', 'Maharashtra', 'Telangana', 'MH', 'Delhi', 'TN', 'West Bengal', 'MH', 'Mumbai', 'Tamil Nadu', 'MH', 'Maharashtra']}
df_visitors_ml = pd.DataFrame(data)

# Count of visitors in each city and state (TOP 10)
df_modified1 = df_visitors_ml.copy()

# Convert city names to uppercase for consistency
df_modified1['city'] = df_modified1['city'].str.upper()

# Drop rows with missing city or state values
df_modified1.dropna(subset=['city', 'state'], inplace=True)

# Define a mapping of patterns to standard city names
pattern_mapping = {
    '.*MUM.*': 'Mumbai',
}

# Apply the pattern matching and replacement
for pattern, replacement in pattern_mapping.items():
    df_modified1.loc[df_modified1['city'].str.contains(pattern, regex=True), 'city'] = replacement

fig, ax = plt.subplots(1,2, figsize=(20,10))
df_modified1['state'].value_counts().head(10).plot(kind='bar', ax=ax[0])
ax[0].set_xlabel('State')
ax[0].set_ylabel('Number of Visitors')
ax[0].set_title('Number of Visitors in Each State')
df_modified1['city'].value_counts().head(10).plot(kind='bar', ax=ax[1])
ax[1].set_xlabel('City')
ax[1].set_ylabel('Number of Visitors')
ax[1].set_title('Number of Visitors in Each City')
plt.show()
```
